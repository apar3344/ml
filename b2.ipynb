```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("drive/MyDrive/MLLAB/glass.csv")
```

```python
df
```

```python
X = df.drop('Type', axis=1)
y = df['Type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Function to train and evaluate SVM with a given kernel
def train_and_evaluate(kernel, X_train, X_test, y_train, y_test):
    # Train the SVM classifier
    svm_classifier = SVC(kernel=kernel, decision_function_shape='ovr')
    svm_classifier.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = svm_classifier.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    return accuracy
```

```python

# Evaluate SVM with RBF kernel
accuracy_rbf = train_and_evaluate('rbf', X_train, X_test, y_train, y_test)
print(f'Accuracy with RBF kernel: {accuracy_rbf:.2f}')
```

```python
# Evaluate SVM with Polynomial kernel
accuracy_poly = train_and_evaluate('poly', X_train, X_test, y_train, y_test)
print(f'Accuracy with Polynomial kernel: {accuracy_poly:.2f}')
```

```python
# Evaluate SVM with Sigmoid kernel
accuracy_sigmoid = train_and_evaluate('sigmoid', X_train, X_test, y_train, y_test)
print(f'Accuracy with Sigmoid kernel: {accuracy_sigmoid:.2f}')
```
