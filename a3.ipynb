```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
file_path = "drive/MyDrive/MLLAB/kmeans.csv"
df = pd.read_csv(file_path)

# Initial cluster centers
mu1 = np.array([6.2, 3.2])
mu2 = np.array([6.6, 3.7])
mu3 = np.array([6.5, 3.0])


# Function to plot the data and cluster centers
def plot_clusters(data, centers, clusters):
    plt.scatter(data[:, 0], data[:, 1], c=clusters, cmap='viridis', edgecolors='k', alpha=0.8)
    plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Cluster Centers')
    plt.title('K-Means Clustering')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.legend()
    plt.show()

# Run k-means for a specified number of iterations



def k_means(X, num_clusters, num_iterations):
    # Randomly initialize centroids
    centroids = np.array([mu1, mu2, mu3])

    for _ in range(num_iterations):
        # Assign each data point to the nearest centroid
        labels = np.argmin(np.linalg.norm(X[:, None] - centroids, axis=2), axis=1)

        # Update centroids based on the mean of the assigned data points
        centroids = np.array([X[labels == i].mean(axis=0) for i in range(num_clusters)])

    return  centroids,labels



# Convert DataFrame to NumPy array
X = df[['X1', 'X2']].values

# Number of clusters (k)
k = 3

# Number of iterations
num_iterations = 10  # You can adjust this based on the convergence criteria

# Run k-means
final_centers, final_clusters = k_means(X,k, num_iterations)

# Plot the results
plot_clusters(X, final_centers, final_clusters)

# Output the results
print("Csenter of the first cluster (red) after one iteration:", np.round(final_centers[0], 3))
print("Center of the second cluster (green) after two iterations:", np.round(final_centers[1], 3))
print("Center of the third cluster (blue) when clustering converges:", np.round(final_centers[2], 3))
print("Number of iterations required for convergence:", num_iterations)
```
