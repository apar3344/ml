```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
file_path = "drive/MyDrive/MLLAB/zoo.csv"
data = pd.read_csv(file_path, header=None)

# Assign columns for better readability
data.columns = ['X', 'y']

# Scatter plot of the data
plt.scatter(data['X'], data['y'])
plt.title('Scatter Plot of the Data')
plt.xlabel('X')
plt.ylabel('y')
plt.show()

# Calculate the correlation matrix
correlation_matrix = data.corr()
print("Correlation Matrix:")
print(correlation_matrix)

# Linear Regression
X = np.column_stack((np.ones(len(data)), data['X']))
y = data['y'].values.reshape(-1, 1)

# Compute regression parameters using the normal equation
theta = np.linalg.inv(X.T @ X) @ X.T @ y

# Predictions
y_pred = X @ theta

# Calculate metrics
n = len(y)
m = 1  # number of features (excluding the intercept term)
degrees_of_freedom = n - m - 1

# Cost function (Mean Squared Error)
cost = np.sum((y_pred - y) ** 2) / (2 * n)

# Sum of Squared Errors (SSE)
sse = np.sum((y_pred - y) ** 2)

# Sum of Squared Residuals (SSR)
ssr = np.sum((y_pred - np.mean(y)) ** 2)

# Total Sum of Squares (SST)
sst = np.sum((y - np.mean(y)) ** 2)

# R-squared (Coefficient of Determination)
r_squared = 1 - (sse / sst)

# Print the results
print("\nLinear Regression Parameters:")
print(f"Intercept (theta0): {theta[0][0]}")
print(f"Slope (theta1): {theta[1][0]}")

print("\nMetrics:")
print(f"Cost: {cost}")
print(f"SSE: {sse}")
print(f"SSR: {ssr}")
print(f"SST: {sst}")
print(f"R-squared (R2): {r_squared}")
```
