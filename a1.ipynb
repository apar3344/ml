```python
import numpy as np

# Given array
data = [115.3, 195.5, 120.5, 110.2, 90.4, 105.6, 110.9, 116.3, 122.3, 110.2]

# Mean
mean_value = np.mean(data)

# Median
median_value = np.median(data)

sorted_data = sorted(data)
mid = len(data)//2
median_valuee = (sorted_data[mid] + sorted_data[mid+1] )/2
print('median : ', median_valuee)
print(len(data))
```

```python
# Mode
def calculate_mode(data):
    unique_values, counts = np.unique(data, return_counts=True)
    max_count_index = np.argmax(counts)
    print(unique_values)
    print(counts)
    print(max_count_index)
    return unique_values[max_count_index]


mode_value = calculate_mode(data)
mode_value
```

```python
# Standard Deviation
variance = sum((x - mean_value) ** 2 for x in data) / len(data)
std_deviation = variance ** 0.5

# Min-Max Normalization
min_value = min(data)
max_value = max(data)
normalized_data = [(x - min_value) / (max_value - min_value) for x in data]

# Standardization (Z-Score)
standardized_data = [(x - mean_value) / std_deviation for x in data]

# Print the results
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
print("Standard Deviation:", std_deviation)
print("Variance:", variance)
print("Min-Max Normalization:", normalized_data)
print("Standardization (Z-Score):", standardized_data)
```
